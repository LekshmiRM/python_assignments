1. Write a Python program that takes an integer as input and calculates the
sum of its digits


```python
sum=0
num=input("Enter a number :")
for i in num :
   sum=sum+int(i)
print("Sum of digits is ", sum)
```

2. Create a Python program to calculate the factorial of a given number using
a function.

```python
def factorial(n) :
   if n==0 :
     return 1
   else :
       return n*factorial(n-1)
num=int(input("Enter a number :"))
print(f"The factorial of  {num} is {factorial(num)}")
```

3. Develop a Python program that checks whether a given word or phrase is a
palindrome.


```python
string=input("Enter a string")
revstr=""
for i in string :
  revstr=i+revstr
print("REversed string is ",revstr)
if(string==revstr) :
  print("The string is palindrome")
else :
    print("The string is not palindrome")
```

4. Create a Python program that takes a sentence as input and counts the
number of vowels and consonants.

```python
sentence="I like python"
vowels = "aeiouAEIOU"
vowel_count = 0
consonant_count = 0

for char in sentence:
   if char.isalpha():
    if char in vowels:
        vowel_count += 1
    else:
        consonant_count += 1
print("Vowels count is :",vowel_count)
print("Consonant count is :",consonant_count)
```

5. Develop a Python program that takes two lists as input and returns a new
list containing the common elements between them.


```python
list1 = ["apple", "orange", "kiwi", "banana"]
list2 = ["orange", "banana", "fig", "dates"]


common = []

for item in list1:
    if item in list2 and item not in common:
        common.append(item)

print("Common elements:", common)
```

6. Write a Python program that takes a string as input and counts the
frequency of each character in the string

```python
text = "I like python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1


for char, count in frequency.items():
    print(char, ":", count)
```

7. Develop a Python program that checks if a given year is a leap year.

```python
year = int(input("Enter a year: "))

if year % 4 == 0:
    print("Leap year")
else:
    print("Not a leap year")
```

10. Write a python program to find the roots of a quadratic equation

```python
import cmath


a = float(input("Enter coefficient a: "))
b = float(input("Enter coefficient b: "))
c = float(input("Enter coefficient c: "))

d = b**2 - 4*a*c


root1 = (-b + cmath.sqrt(d)) / (2*a)
root2 = (-b - cmath.sqrt(d)) / (2*a)
print("The roots of the quadratic equation are:")
print("Root 1 =", root1)
print("Root 2 =", root2)
```
