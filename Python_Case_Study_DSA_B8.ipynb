# Q1. Write a function that return the range (maximum and minimun) of a given list of numbers

```python
def find_range(list_numbers):
  minimum_value=list_numbers[0] #setting the first value of the list as minimum value
  maximum_value=list_numbers[0] #setting the first value of the list as minimum value

  for number in list_numbers :  #looping through the entire list to check for maximum_value and minimum_value
    if(number>=maximum_value):
     maximum_value=number
    if(number<=minimum_value):
     minimum_value=number

  return(minimum_value,maximum_value)

mylist=[34,99,14,87,8,49,6,3]
range=find_range(mylist)
range


```

## Q2. Write function that returns mean value of a list of numbers

```python
def mean(list_number):
  sum=0
  for num in list_number:
    sum+=num
  return(sum/len(list_number))

list_number=[1 ,2 ,3 ,4 ,5]
#mean(list_number)
print("Mean value of list is ", mean(list_number))
```
